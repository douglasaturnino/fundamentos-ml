# 0.0 Imports

```python
import cv2
from sklearn import datasets as dt
from sklearn import tree as tr
from matplotlib import pyplot as plt
```

# 1.0  Load dataset

```python
iris = dt.load_iris()

X = iris.data[:,2:]
y = iris.target
```

# 2.0  Model Training

```python
# difine
model_tree = tr.DecisionTreeClassifier()

# traininf (fit)
model_tree.fit(X, y)

# predict
```

```python
tr.export_graphviz(
    model_tree,
    out_file='tree.dot',
    feature_names=iris.feature_names[2:],
    class_names=iris.target_names,
    rounded=True,
    filled=True
)
```
